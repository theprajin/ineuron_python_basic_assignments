### 1. Why are functions advantageous to have in your programs?

While programming, at times, we encounter that we have to write same piece of code repeatedly at different places to get the same result. So, in this case it is better to write that piece of code as a function so that it can be called whenever required. This technique increases redundancy and reusability of a program. Also, function helps to improve code clearity as we don't have to write all the code in one place in cluster by the use of function.

### 5. How many global scopes are there in a Python program? How many local scopes?

### 6. What happens to variables in a local scope when the function call returns?

### 7. What is the concept of a return value? Is it possible to have a return value in an expression?

The return value is the value that is returned to a function caller while the "return" keyword marks the end of a function execution and codes after return statement are not executed. It is not possible to have a return value in an expression. It can only used inside a function.

### 8. If a function does not have a return statement, what is the return value of a call to that function?
