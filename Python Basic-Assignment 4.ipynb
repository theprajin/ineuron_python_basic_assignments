### 15. How do you get a list value's tuple form? How do you get a tuple value's list form?

To get a list's tuple form we use the built-in method tuple() and to get a tuple value's list form we use the method list(). This technique is also called type-casting, changing from one data type to another.

```python
list1 = ['prajin', 'bajracharya', 134,34, True, 56.2]
tuple1 = ('python', 234, False, 345.6, "numpy", "pandas")
```

```python
#changing from list to tuple
tuple_from_list = tuple(list1)
print(tuple_from_list)
type(tuple_from_list)
```

```python
#changing from tuple to list
list_from_tuple = list(tuple1)
print(list_from_tuple)
type(list_from_tuple)
```

### 17. How do you distinguish between copy.copy() and copy.deepcopy()?

The copy.copy() will do the shallow copy only. Meaning when copying it doesn't create entirely new object but instead make references to the item of the original object. On the contrary, the copy.deepcopy() will do the deep copy of an object and creates an entirely new object.

```python
import copy

#Demo of shallow copy using copy.copy()
l1 = [1,2,4, [5,6,7], 4,5,6]
l2 = copy.copy(l1)

print("The original list before shallow copying")
print(l1)

print("The list after shallow copying with modification of the list")
l2[3][1] = 15694
print(l2)
print(l1)
```

```python
#Demo of deep copy using copy.copy()
l1 = [1,2,4, [5,6,7], 4,5,6]
l2 = copy.deepcopy(l1)

print("The original list before shallow copying")
print(l1)

print("The list after shallow copying with modification of the list")
l2[3][1] = 15694
print(l2)
print(l1)
```
