### 3. What is the most significant distinction between a dictionary and a list?

The most significant distinction between a dictionary and a list is that list is accessed using index starting from 0 and dictionary is accessed using key given to each value. Also, unlike lists items in a dictionary is unordered.

```python
lst = [0,1,2,3,4]
dct = {'zero':0, 'one':1, 'two':2, 'three':3, 'four':4}

print("Three from list: {}".format(lst[3]))
print("Three from dictionary: {}".format(dct['three']))
```

### 7. What is a shortcut for the following code?
### if 'color' not in spam:
### spam['color'] = 'black'

The shortcut for the code block given above is to use the built-in setdefault() method of the dictionary.

```python
spam = {'fruit': 'apple', 'taste': 'sweet'}
spam.setdefault('color', 'black')
print(spam)
```

### 8. How do you "pretty print" dictionary values using which module and function?

This can be done by using the pprint function of print module.

```python
import pprint
spam = {
    "person1":{'name': 'Joseph', 'age':23,},
    "person2":{'name': 'Guardiola', 'age':45,},
    "person3":{'name': 'Nancy', 'age':13,},
    "person4":{'name': 'Juan', 'age':54,},
    "person5":{'name': 'Mikel', 'age':87,}
}
```
