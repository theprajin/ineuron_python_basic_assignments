### 1. What are escape characters, and how do you use them?

Escape characters are the ones that allow to include special characters in a string. To do this, a backslash is added the character that is to be escaped.

### 2. What do the escape characters n and t stand for?

The escape character n stands for leave a line and t stands for leave a tab.

```python
print("stringisthebestone.\nand\tso is the list")
```

### 9. What are the methods for right-justifying, left-justifying, and centering a string?

The methods for right-justifying, left-justifying, and centering a string are rjust(), ljust(), and center() respectively.

```python
s = 'lkasdjf'
```

```python
print(s.ljust(20,'o'))
```

```python
print(s.rjust(12,'r'))
```

```python
print(s.center(25,'e'))
```

### 10. What is the best way to remove whitespace characters from the start or end?

The string methods strip can bbe used to remove whitespace character from both start and end of a string literal. And the string methods lstrip and rstrip is used to remove whitespace character from start and end of a string respectively.

```python
"   Removing the whitespace characters from the start or end  ".strip()
```

```python
"   Removing the whitespace characters from the start or end  ".lstrip()
```

```python
"   Removing the whitespace characters from the start or end  ".rstrip()
```
