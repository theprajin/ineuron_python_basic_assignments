### 1. To what does a relative path refer?

The relative path is the path relative to the current directory.

### 2. What does an absolute path start with your operating system?

Since my operating system is Windows so, the absolute path starts from C:\.

### 3. What do the functions os.getcwd() and os.chdir() do?

The os.getcwd() returns the current working directory and the os.chdir() changes the current directory to another directory.

### 4. What are the . and .. folders?

The . folder is the current directory and the .. folder is the parent directory.

### 5. In C:\bacon\eggs\spam.txt, which part is the dir name, and which part is the base name?

Here, C:\bacon\eggs is the directory and spam.txt is the base name.

### 6. What are the three “mode” arguments that can be passed to the open() function?

The three mode arguments that can be passed to the open() function are 'r', 'w', 'a' which refers to read, write, and append mode respectively

### 7. What happens if an existing file is opened in write mode?

The contents in the existing is completely rewritten.

### 8. How do you tell the difference between read() and readlines()?

The read() will read an entire file and return in a form of a string if n byte is not specified as its parameter. If n byte is specified, reads the byte and return as string. And the readlines() reads the entire file and return each line of string as list.
